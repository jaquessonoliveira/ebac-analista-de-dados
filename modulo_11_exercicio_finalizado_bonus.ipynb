```python
def crawl_website(url: str) -> str:
  
  try:
    resposta = requests.get(url=URL)
    resposta.raise_for_status()
  except HTTPError as exc:
    print(exc)
  else:
    return resposta.text
```

```python
pagina = BeautifulSoup(open('trending.html', mode='r'), 'html.parser')
```

```python
# Desenvolvendo o exercício através da correção:

tabela = pagina.find_all('article')

project = []
language = []
stars = []
stars_today = []
forks = []

for item in tabela:
  # Nome do projeto:
  project.append(item.find('h1',{'class': 'h3 lh-condensed'}).get_text().strip().replace('\n','').replace(' ',''))
  # Linguagem utilizada:
  linguagem = item.find('span',{'itemprop': 'programmingLanguage'})
  try:
    linguagem = linguagem.get_text()
  except AttributeError:
    linguagem = ''
  language.append(linguagem)
  # Stars e Forks recebidos:
  conjunto = item.find_all('a',{'class': 'Link--muted d-inline-block mr-3'})
  stars.append(conjunto[0].get_text().replace(',','').strip())
  forks.append(conjunto[1].get_text().replace(',','').strip())
  # Stars_today recebidos:
  stars_today.append(item.find('span',{'class': 'd-inline-block float-sm-right'}).get_text().replace('stars today','').strip())

project = project[0:10]
language = language[0:10]
stars = stars[0:10]
stars_today = stars_today[0:10]
forks = forks[0:10]

ranking = []
for i in range(1, 11):
  ranking.append(i)

```

```python
import csv
import pandas as pd

with open(file='github.csv', mode='w', encoding='utf8') as fp:
   escritor_csv = csv.writer(fp, delimiter=';')
   escritor_csv.writerow(['ranking','project','language','stars','stars_today','forks'])
   for i in range(len(ranking)):
     escritor_csv.writerow([ranking[i],project[i],language[i],stars[i],stars_today[i],forks[i]])

print(escritor_csv)
```
